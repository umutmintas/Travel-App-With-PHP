## Web site Features

# Features
- G Translate kullanarak tüm sayfanın 12 farklı dilde anlaşılabilmesini sağladım. Web sitesine giriş yapan kullanıcının tarayıcı dili ne ise ona göre sayfayı otomatik olarak çeviriyor.  
- web sitesinde bir de yüzen whatsapp butonu bulunmakta buda firma ile hızlı iletişim kurulabilmesi için önemli özelliklerden birtanesi


   # Home Page
-  Anasayfada vitrin tasarmına benzer bir yapı kurmak istedim ve popüler tur hizmetlerini orada sergilerdim 

   # About Page
-  Hakkımızda alanında firmayı detaylı olarak açıklaycak şekilde bilgileri girdim ve tüm himetlerini detaylı olarak açıkladım

   # Tour Page
-  Tur sayfasında tüm turları sergiledim 

   # Rezervation Page
- Kullanıcıların firma ile iletişime geçebileceği birçok seçenek ekledim
- Kullanıcıların bir tur programına katılabilmesi veya bilgi alabilmesi için bir form oluşturdum
